## KMeans Clustering

```python
air_data=pd.read_excel(r"C:\Users\Tejasvi\Downloads\EastWestAirlines.xlsx")
```

```python
air1_data=air_data.drop(['ID#'],axis=1)
air1_data
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df2_norm = norm_func(air1_data)
df2_norm
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df2_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above plot
model1=KMeans(n_clusters=4)
model1.fit(df2_norm)
```

```python
model1.labels_
```

```python
#Assign clusters to the data set
air1_data['clusterid_new'] = model1.labels_
air1_data
```

```python
model1.cluster_centers_
```

```python
air1_data.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10,7))  
plt.scatter(air1_data['clusterid_new'],air1_data['Balance'], c=model1.labels_)
```
